```python
import os
import random as rd

# -------------------------------------------------------
# Step 1: Initialization & File Creation
# -------------------------------------------------------
# Create the target directory "ass2"
# Note: This will raise FileExistsError if folder already exists
os.makedirs("ass2"\) 

# Loop to create 10 dummy files
for i in range(10):
    # 'w' mode creates the file if it doesn't exist
    open(f"ass2/file_{i}.txt", "w").close()

# -------------------------------------------------------
# Step 2: Listing & Random Selection
# -------------------------------------------------------
# Get list of all filenames in the folder
all_files = os.listdir("ass2")
all_count = len(all_files)

# Select random 50% of the files
# len(all_files)//2 ensures we get an integer number (e.g., 5)
selected_files = rd.sample(all_files, len(all_files)//2)

# -------------------------------------------------------
# Step 3: Deletion (The Snap)
# -------------------------------------------------------
for file in selected_files:
    # Construct the full path (e.g., "ass2/file_1.txt")
    file_path = os.path.join("ass2", file)
    
    # Remove the file
    os.remove(file_path)

# -------------------------------------------------------
# Step 4: Verification
# -------------------------------------------------------
# Check directory content after deletion
remaining_files = os.listdir("ass2")

print(f"Total files created: {all_count}")
print(f"Files remaining after deletion: {len(remaining_files)}")
```
