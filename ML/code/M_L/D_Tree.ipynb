```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
data=pd.read_csv(r'D:\AI\ai_deploma_master\ai_deploma_master\files\PlayTennis 1.csv')
```

```python
data.head()
```

```python
data.columns
data.columns[0]
```

```python
from sklearn.preprocessing import OrdinalEncoder
for i in data.columns:
    encoder = OrdinalEncoder()
    data[i] = encoder.fit_transform(data[[i]])
```

```python
outlook_encoder=OrdinalEncoder()
data['Outlook']=outlook_encoder.fit_transform(data[['Outlook']])
```

```python
temp_encoder=OrdinalEncoder()
data['Temperature']=temp_encoder.fit_transform(data[['Temperature']])
```

```python
Hum_encoder=OrdinalEncoder()
data['Humidity']=Hum_encoder.fit_transform(data[['Humidity']])
```

```python
Wind_encoder=OrdinalEncoder()
data['Wind']=Wind_encoder.fit_transform(data[['Wind']])
```

```python
Play_Tennis_encoder=OrdinalEncoder()
data['Play Tennis']=Play_Tennis_encoder.fit_transform(data[['Play Tennis']])
```

```python
data
```

```python
x=data.drop(['Play Tennis'],axis=1)
y=data['Play Tennis']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.8,random_state = 10)
```

```python
clf=DecisionTreeClassifier(criterion='gini')
clf.fit(x_train, y_train)
y_pred=clf.predict( x_test )
```

```python
clf_1=DecisionTreeClassifier(criterion='entropy')
clf_1.fit(x_train, y_train)
y_pred_1=clf.predict( x_test )
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("accuracy with gini",accuracy_score(y_test,y_pred))
print('--------------------------------------------------')
print("accuracy with entorpy",accuracy_score(y_test,y_pred_1))
print('--------------------------------------------------')
print("classification_report",classification_report(y_test,y_pred))
print('--------------------------------------------------')
print("classification_report",classification_report(y_test,y_pred_1))
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf,filled=True,feature_names=x.columns,class_names=Play_Tennis_encoder.categories_[0],rounded=True)
plt.show()
```
