```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Salary_Data.csv")
plt.Figure(figsize=(3,3))
# plt.scatter(df.YearsExperience,df.Salary,color="red")
# plt.xlabel="years"
# plt.ylabel="salary"
sns.pairplot(df,x_vars=["YearsExperience"],y_vars=["Salary"],size=7,kind="scatter")
plt.xlabel="years"
plt.ylabel="salary"
df
```

```python
x = df.iloc[:, :-1] # كل الصفوف، وكل الأعمدة ما عدا الأخير
y = df.iloc[:, -1]  # كل الصفوف، والعمود الأخير فقط
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
liner_model=LinearRegression()
liner_model.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,liner_model.predict(x_train),color="red")
plt.xlabel="years"
plt.ylabel="salary"
plt.title("training data")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_test,liner_model.predict(x_test),color="red")
plt.xlabel="years"
plt.ylabel="salary"
plt.title("test data")
plt.show()
```

```python
y_pred=liner_model.predict(x_test)
y_pred
```

```python
# calculating the mean squared error and r2 score
mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)
rms=root_mean_squared_error(y_test,y_pred)
print("Root Mean Squared Error:",rms)#(mse)**0.5
r2=r2_score(y_test,y_pred)
print("R2 Score:",r2)
```

```python
b=liner_model.intercept_
```

```python
w=float(liner_model.coef_) # a
print("w=",w)
print("b=",b)
```
