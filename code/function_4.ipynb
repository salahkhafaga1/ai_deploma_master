```python
def user_info(name,age):
     """
    Takes the user's name and age from the user interface
    and prints them.

    :param name: The name of the user.
    :type name: str

    :param age: The age of the user.
    :type age: int

    :return: None
    """
    return f"Name: {name}, Age: {age}"
```
