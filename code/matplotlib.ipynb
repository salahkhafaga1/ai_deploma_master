```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('D:/ai_deploma_master/ai_deploma_master/code/files/train (2).csv')
cat_cols=df.select_dtypes('category').columns
```

```python
plt.figure(figsize=(14,4))
for i ,col in enumerate(cat_cols):
    plt.subplot(2,3,i+1)
    sns.countplot(x=col,data=df)
    plt.title(f"{col} Distrib. Graph" )
plt.subplots_adjust(hspace=.8,wspace=0.3)
plt.show()
# plt.Figure()

```

```python
plt.figure(figsize=(9,4))
for i ,col in enumerate(cat_cols):
    plt.subplot(2,3,i+1)
    unique=df['Survived'].value_counts()
    count=unique.values
    catogrey=unique.index
    plt.pie(count,labels=catogrey,startangle=1440,autopct='%1.1d%%')
    plt.title(f"{col} Pie. Graph" )

plt.subplots_adjust(hspace=.8,wspace=0.3)
plt.show()
```
