```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('D:/ai_deploma_master/ai_deploma_master/code/files/train (2).csv')
cat_cols=df.select_dtypes('category').columns
```

```python
plt.figure(figsize=(14,4))
for i ,col in enumerate(cat_cols):
    plt.subplot(2,3,i+1)
    sns.countplot(x=col,data=df)
    plt.title(f"{col} Distrib. Graph" )
plt.subplots_adjust(hspace=.8,wspace=0.3)
plt.show()
# plt.Figure()

```

```python
plt.figure(figsize=(9,4))
for i ,col in enumerate(cat_cols):
    plt.subplot(2,3,i+1)
    unique=df['Survived'].value_counts()
    count=unique.values
    catogrey=unique.index
    plt.pie(count,labels=catogrey,startangle=1440,autopct='%1.1d%%')
    plt.title(f"{col} Pie. Graph" )

plt.subplots_adjust(hspace=.8,wspace=0.3)
plt.show()
```

```python
xpoints = np.array([1, 8,9])
ypoints = np.array([3, 10,8])

plt.plot(xpoints, ypoints,marker = 'o')
plt.plot(ypoints,marker = 'o')
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
# Markers
```

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()
```

**Matplotlib Subplot**
Display Multiple Plots
With the subplot() function you can draw multiple plots in one figure:

```python
#plot 1:
x = np.array([0, 1, 2, 3,4])
y = np.array([3, 8, 1, 10,5])

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.

plt.plot(x,y)

plt.show()
```

**Matplotlib Bars**

```python
# Matplotlib Bars
# barh()
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
# verticalbar
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```
