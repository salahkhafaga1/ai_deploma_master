```python
# 1. تعريف الكلاس (Class Definition)
class student:
    # هنا حط البيانات (Attributes) بقيم ثابتة مباشرة جوه الكلاس
    student_name = "amit learning"
    student_age = 10
    student_GPA = 3.99
    student_gender = "Male"

    # دي دالة (Method) - (ملاحظة: في الصورة مكتوبة للتوضيح فقط)
    def student_info(self):
        print(f"student name: {self.student_name}")

# ---------------------------------------------------------
# الخطوات اللي تحت في الصورة (التطبيق العملي)
# ---------------------------------------------------------

# 2. إنشاء كائن (Create Object / Instance)
# السهم في الصورة بيقول: Copy from Class -> Create Object
s1 = student()
s1.student_info()
# 3. الوصول للبيانات (Access Attribute)
# السهم الأخير بيقول: Access Attribute student name
print(s1.student_name)
# أي دالة (Method) جوه الكلاس، عشان تشوف البيانات (Attributes) اللي معاها في نفس الكلاس، لازم نستخدم كلمة self. قبل اسم البيانات.
```

```python
# How To Access Method
'''In Python's object-oriented programming paradigm, instance methods within a class conventionally 
include a reference to the instance itself through the first parameter. This allows the method to access'''
'''The use of self in Python's object-oriented programming serves as a convention to refer to the instance of 
the class within its methods. When you define a method inside a class and include self as its first parameter, 
you are essentially telling Python that this method is meant to operate on an instance of the class

'''
class student:
    student_name = "amit learning"

    # ---------------------------------------------------------
    # قاعدة الـ self (أهم قاعدة في الميثود)
    # ---------------------------------------------------------
    # 1. أي دالة جوه الكلاس لازم تاخد باراميتر أول (بنسميه self).
    # 2. كلمة self هي مجرد "إشارة" للكائن اللي بينادي الدالة دلوقتي.
    # 3. بايثون بيمرر الكائن (s1) لمكان (self) بشكل أوتوماتيك مخفي.
    
    def student_info(self):
        # self هنا شايلة جواها s1 بكل بياناته
        print("hello world")

s1 = student()

# اللحظة دي بيحصل فيها السحر:
# s1.student_info()  <--- بتتحول في الخلفية لـ ---> student_info(s1)
s1.student_info()
```

```python
# Prove that self variable is a reference to the instance of the class
# ---------------------------------------------------------
# تجربة عملية: إثبات أن self هي نفسها الكائن (Instance)
# ---------------------------------------------------------

class test():
    def info(self):
        # هنا بنطبع قيمة self لنرى عنوانها في الذاكرة
        print(f"self refers to object at address: {self}")

# 1. إنشاء كائن (Instance)
t1 = test()

# 2. طباعة الكائن من الخارج
print(f"t1 refers to object at address:   {t1}")

# 3. استدعاء الدالة (التي ستطبع self من الداخل)
t1.info()

# --- الخلاصة (Conclusion) ---
# ستلاحظ أن العنوان (Memory Address) المطبوع في الحالتين متطابق تماماً.
# هذا يثبت أن:
# المتغير (t1) خارج الكلاس = المتغير (self) داخل الكلاس.
# self is basically a reference to the current instance being manipulated.
```

```python
# Object Oriented Programming Language
# --- قاموس مفاهيم OOP (OOP Concepts Dictionary) ---

'''1. Constructor (__init__):
   - دالة خاصة تعمل تلقائياً عند إنشاء الكائن.
   - وظيفتها: تحديد القيم الأولية للخصائص (Setting initial state).

2. Inheritance (الوراثة):
   - قدرة الكلاس الجديد على اكتساب صفات ودوال من كلاس قديم.
   - الفائدة: إعادة استخدام الكود (Code Reuse) وتنظيمه.
   - نستخدم دالة super() للوصول لخصائص الأب.

3. Encapsulation (التغليف):
   - تجميع البيانات والدوال في وحدة واحدة (Class).
   - الهدف: إخفاء التفاصيل الداخلية وحماية البيانات من التعديل الخطأ (Controlling access).

4. Polymorphism (تعدد الأشكال):
   - استخدام نفس اسم الدالة لأغراض مختلفة (نفس الفعل بس بطرق تنفيذ مختلفة).
   - مثال: Method Overriding (تغيير وظيفة دالة ورثناها من الأب).

5. Destructor (__del__):
   - دالة تعمل تلقائياً قبل تدمير الكائن ومسحه من الذاكرة.
   - الفائدة: تنظيف الموارد (Cleanup) ولكن يفضل عدم الاعتماد عليها كلياً في بايثون.

6. Multiple Inheritance (الوراثة المتعددة):
   - أن يرث كلاس واحد من أكثر من كلاس أب في نفس الوقت.
   - ملاحظة: تتطلب ترتيب معين للأولويات يسمى (MRO - Method Resolution Order).'''
```

```python
# 1. Constructor (__init__)
'''In Python, a constructor is like a helper that sets up an object automatically when you create it. It uses a 
special method called __init__ to give the object its starting values and make sure it begins in a specific state'''
class student():
    """
    مفهوم الـ Constructor (البنّاء):
    دالة __init__ هي دالة خاصة جداً.
    لا نحتاج لاستدعائها يدوياً (مثل s1.init()).
    بل يتم استدعاؤها تلقائياً (Automatically invoked) لحظة إنشاء الكائن.
    """
    
    # method (Constructor)
    def __init__(self):
        # هذا السطر سيعمل فوراً عند كتابة s1 = student()
        # الغرض: عادة نستخدم هذا المكان لوضع القيم الأولية، أو رسائل الترحيب.
        print("Hello from constructor")

# --- التنفيذ ---
# بمجرد تنفيذ هذا السطر، ستظهر رسالة الطباعة
s1 = student()
```

```python
# 2. Abstraction
'''Abstraction in python is defined as a process of handling complexity by hiding unnecessary information 
from the user.'''
# استيراد مكتبة الـ ABC (Abstract Base Class)
# دي المكتبة المسؤولة عن وضع "القواعد والقوانين" للكلاسات
from abc import ABC, abstractmethod

# 1. تعريف كلاس مجرد (Abstract Class)
# هذا الكلاس هو مجرد "فكرة" أو "تصميم".
# هام جداً: لا يمكن إنشاء كائن (Object) من هذا الكلاس مباشرة.
# يعني مينفعش نكتب: x = Animal() <-- هيعطي خطأ
class Animal(ABC):
    
    # 2. تعريف دالة مجردة (Abstract Method)
    # العلامة @abstractmethod معناها:
    # "يا أي كلاس يورث مني، أنت مُجبر (Must) تعمل الدالة دي عندك".
    @abstractmethod
    def move(self):
        pass  # هنا الدالة فاضية لأن الأب ميعرفش الابن هيتحرك إزاي

# 3. الكلاسات المشتقة (Concrete Classes)
# بمجرد ما ورثنا من Animal، أصبح لازم نعمل دالة move
class Brid(Animal):
    def move(self):
        print("Flying")  # هنا التنفيذ الفعلي (Implementation)

class Cat(Animal):
    def move(self):
        print("Meow")

# --- التجربة ---
# إنشاء كائن من الكلاس الابن (مسموح)
a = Brid()
a.move()
b = Cat()
b.move()
'''Abstraction in python is defined as a process of handling complexity by hiding unnecessary information 
from the user'''
    
```

```python
# error Example of Abstraction
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def move(self):
        pass

# ---------------------------------------------------------
# الحالة الخاطئة (The Trap)
# ---------------------------------------------------------
class Bird(Animal):
    # هذا الكلاس ورث الدالة المجردة move ولكنه لم يقم بتنفيذها (Implementation).
    # النتيجة: يعتبر بايثون هذا الكلاس أيضاً Abstract Class.
    pass 

# محاولة إنشاء كائن
# a = Bird()  <-- (Error)
# السبب: لا يمكن إنشاء كائن من كلاس ورث دالة مجردة ولم يقم بتنفيذها.

# ---------------------------------------------------------
# الحالة الصحيحة (The Fix)
# ---------------------------------------------------------
class Cat(Animal):
    # هذا الكلاس نفذ الشرط وقام بكتابة الكود الخاص بالدالة.
    def move(self):
        print("move from Cat")

c = Cat()
 # (Correct) - لأن الكلاس أصبح مكتملاً.
c.move()    
```

```python
# 3. Encapsulation
'''Encapsulation is the practice of hiding the internal data of a class and exposing only what is necessary for the outside 
world to interact with the class.'''
# there 2 type in attributes:
# 1. Public Attribute:Private variables and methods are those that can only be accessed within the class.
# 2. Private Attribute:Public variables and methods are those that can be accessed by any part of the program. 
# ----------
# private attribute:Private variables and methods are those that can only be accessed within the class.
class Car:
    # 1. تعريف متغير خاص (Private Attribute)
    # البادئة (__) تجعل المتغير غير مرئي خارج الكلاس.
    __engine_capacity = "2000cc"

    # 2. تعريف دالة خاصة (Private Method)
    # لا يمكن استدعاؤها من الخارج بـ c1.__start_engine()
    def __start_engine(self):
        print("Engine Started")

# --- محاولة الوصول (Access Attempt) ---
c1 = Car()

# الخطأ (The Error):
print(c1.__engine_capacity)
# النتيجة: AttributeError
# السبب: بايثون تقوم بعملية (Name Mangling) لتغيير اسم المتغير داخلياً ومنع الوصول المباشر له.

```

```python
# Public Attribute
class Car:
    # 1. تعريف متغير عام (Public Attribute)
    _engine_capacity_ = "2000cc"

    # 2. تعريف دالة عامة (Public Method)
    def start_engine(self):
        print("Engine Started")

# --- الوصول (Access) ---
c1 = Car()
# الصحيح (The Correct Way):
print(c1._engine_capacity_)
```

```python
# 4. Polymorphism
'''Polymorphism is a concept in object-oriented programming that allows objects of different types to be 
treated as objects of a common type. '''
# 1. الكلاس الأب (الشكل العام)
class Shape:
    def calculate_area(self):
        pass  # هنا الدالة فاضية، لأن كلمة "شكل" ملهاش مساحة محددة

# 2. الكلاس الابن الأول (دائرة)
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    # إعادة صياغة الدالة (Method Overriding)
    # هنا الدائرة بتقول: "أنا بحسب المساحة بطريقتي الخاصة (ط نق تربيع)"
    def calculate_area(self):
        return 3.14 * (self.radius ** 2)

# 3. الكلاس الابن الثاني (مربع)
class Square(Shape):
    def __init__(self, side):
        self.side = side

    # إعادة صياغة الدالة مرة تانية
    # هنا المربع بيقول: "وأنا كمان بحسب المساحة بطريقتي (طول الضلع في نفسه)"
    def calculate_area(self):
        return self.side ** 2
    

# --- الاستخدام (السحر هنا) ---

x = Circle(5)
y = Square(4)

# لاحظ: بننادي نفس الدالة 'calculate_area' على الاتنين
# بس النتيجة مختلفة تماماً حسب نوع الكائن
print("Area: ", x.calculate_area())  # هيحسب مساحة الدائرة (78.5)
print("Area: ", y.calculate_area())  # هيحسب مساحة المربع (16)
```

```python
# 5. Inheritance
'''Inheritance is a fundamental concept in object-oriented programming (OOP) where a new class 
(subclass/derived class) can inherit attributes and behaviors from an existing class (superclass/base class).'''
# 1. Parent Class (Super Class) - الكلاس الأب
class Calculator1:
    def summation(self, n1, n2):
        return n1 + n2
    
    def subtraction(self, n1, n2):
        return n1 - n2

# 2. Child Class (Sub Class) - الكلاس الابن
# يرث من Calculator1 عن طريق وضعه بين القوسين
class Calculator2(Calculator1):
    # هذا الكلاس يمتلك الآن دالتي الجمع والطرح تلقائياً (ورثهم)
    
    # ويمكننا إضافة دوال جديدة خاصة به
    def power(self, n1, n2):
        return n1 ** n2

# --- الاستخدام ---
c1 = Calculator2()

# استدعاء دالة موروثة (موجودة عند الأب)
print(c1.summation(2, 5))  # النتيجة: 7

# استدعاء دالة جديدة (موجودة عند الابن)
print(c1.power(2, 3))      # النتيجة: 8
```

```python
# 6. Multiple Inheritance
'''Multiple Inheritance is a concept in object-oriented programming where a class can inherit attributes and 
behaviors from more than one parent class.'''

# --- مفهوم الوراثة المتعددة (Multiple Inheritance) ---

class A:
    def do_this(self):
        print("I am in A")

class B(A):
    pass

class C:
    def do_this(self):
        print("I am in C")

# الترتيب هنا حاسم جداً!
# D(B, C) تعني: ابحث في B (وعائلته) أولاً، ثم ابحث في C.
class D(B, C):
    pass

# --- قاعدة MRO (Method Resolution Order) ---
# 1. بايثون تبحث عن الدالة في الكلاس الحالي (D).
# 2. إذا لم تجدها، تذهب لأول كلاس في القوسين (B).
# 3. إذا لم تجدها في B، تصعد لأب B (وهو A).
# 4. وجدتها في A -> تنفذها فوراً وتتجاهل C.

x = D()
x.do_this() # النتيجة: I am in A
```

```python
# ex2
# --- مفهوم الوراثة المتعددة (Multiple Inheritance) ---

class A:
    def do_this(self):
        print("I am in A")

class B(A):
    pass

class C():
    def do_this(self):
        print("I am in C")

# الترتيب هنا حاسم جداً!
# D(B, C) تعني: ابحث في B (وعائلته) أولاً، ثم ابحث في C.
class D(C, B):
    pass

# --- قاعدة MRO (Method Resolution Order) ---
# 1. بايثون تبحث عن الدالة في الكلاس الحالي (D).
# 2. إذا لم تجدها، تذهب لأول كلاس في القوسين (B).
# 3. إذا لم تجدها في B، تصعد لأب B (وهو A).
# 4. وجدتها في A -> تنفذها فوراً وتتجاهل C.

x = D()
x.do_this() # النتيجة: I am in C
```
