```python
name="salah"
print(name[0])
print(name[-1])
```

```python
print(name[1:4])
print(name[:3])
print(name[2:])
print(name[:])
print(name[-4:-1])
print(name[-3:])
print(name[:-2])
print(name[::-1])
print(name[::-2])
print(name[::2])
print(name[1::2])
```

```python
print(name.upper())
print(name.lower())
print(name.capitalize())
print(name.replace('s', 'S'))
print(name.split('a'))
print(name.find('l'))
print(len(name))
print(name.count('a'))
print(name.startswith('s'))
print(name.endswith('h'))
print(name.isalpha())
print(name.isdigit())
print(name.swapcase())
print(name.find('l'))
```

```python
# join function used to concatinate list of strings into a single string
word = ["sa","ofdk", "asldo", "ofdk", "asldo"]
sentence = ' '.join(word)
print(sentence)
```

```python
# startwith function used to check if a string starts with a specific substring
text = "Hello, welcome to the world of Python."
print(text.startswith("Hello"))
```

```python
# endswith function used to check if a string ends with a specific substring
print(text.endswith("."))
```
